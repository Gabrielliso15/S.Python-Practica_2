```python
import sys
import os

# Agrega el directorio ra√≠z "Entregas" al path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
import src.functions as funct


rounds = [
 {
 'Shadow': {'kills': 2, 'assists': 1, 'deaths': True},
 'Blaze': {'kills': 1, 'assists': 0, 'deaths': False},
 'Viper': {'kills': 1, 'assists': 2, 'deaths': True},
 'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 0, 'assists': 2, 'deaths': False},
 'Blaze': {'kills': 2, 'assists': 0, 'deaths': True},
 'Viper': {'kills': 1, 'assists': 1, 'deaths': False},
 'Frost': {'kills': 2, 'assists': 1, 'deaths': True},
 'Reaper': {'kills': 0, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 1, 'assists': 0, 'deaths': False},
 'Blaze': {'kills': 2, 'assists': 2, 'deaths': True},
 'Viper': {'kills': 1, 'assists': 1, 'deaths': True},
 'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 2, 'assists': 1, 'deaths': False},
 'Blaze': {'kills': 1, 'assists': 0, 'deaths': True},
 'Viper': {'kills': 0, 'assists': 2, 'deaths': False},
 'Frost': {'kills': 1, 'assists': 1, 'deaths': True},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
 },
 {
 'Shadow': {'kills': 1, 'assists': 2, 'deaths': True},
 'Blaze': {'kills': 0, 'assists': 1, 'deaths': False},
 'Viper': {'kills': 2, 'assists': 0, 'deaths': True},
 'Frost': {'kills': 1, 'assists': 1, 'deaths': False},
 'Reaper': {'kills': 1, 'assists': 1, 'deaths': True}
 }
 ]


tabla_final ={
 'Shadow': {'kills': 0, 'assists': 0, 'deaths': 0, 'puntos': 0, 'MVP': 0},
 'Blaze': {'kills': 0, 'assists': 0, 'deaths': 0, 'puntos': 0, 'MVP': 0},
 'Viper': {'kills': 0, 'assists': 0, 'deaths': 0, 'puntos': 0, 'MVP': 0},
 'Frost': {'kills': 0, 'assists': 0, 'deaths': 0, 'puntos': 0, 'MVP': 0},
 'Reaper': {'kills': 0, 'assists': 0, 'deaths': 0, 'puntos': 0, 'MVP': 0}
 }

for i, ronda in enumerate(rounds):
    print()
    print(f"Ronda {i+1}")
    print("Jugador   Kills    Asistencias    Muertes    Puntos")
    print("-------------------------------------------------------")

    puntajes = {}  # diccionario con los puntajes 

    for jugador, stats in ronda.items():
        puntos = funct.calcular_puntos(stats)
        puntajes[jugador] = puntos  # Guardar puntaje 

        murio = "Si" if stats["deaths"] else "No"
        print(f"{jugador:<10} {stats['kills']:^7} {stats['assists']:^12} {murio:^7} {puntos:^7}")

        tabla_final[jugador]["kills"] += stats["kills"]
        tabla_final[jugador]["assists"] += stats["assists"]
        tabla_final[jugador]["deaths"] += 1 if stats["deaths"] else 0  
        tabla_final[jugador]["puntos"] += puntos 

        # Guardar puntaje para el MVP
        puntajes[jugador] = puntos

    # MVP de la ronda
    mvp = funct.Buscar_max(puntajes)[0]
    tabla_final[mvp]["MVP"] += 1
    print("\nMVP de la ronda:", mvp, f"con {puntajes[mvp]} puntos.")
    print("-------------------------------------------------------")


print("\nTabla Final:")
print("Jugador   Kills    Asistencias    Muertes    MVPs    Puntos")
print("---------------------------------------------------------------")
tabla_ordenada = sorted(tabla_final.items(), key=lambda x: x[1]['puntos'], reverse=True)
for jugador, stats in tabla_ordenada:
    print(f"{jugador:<10} {stats['kills']:^7} {stats['assists']:^12} {stats['deaths']:^6} {stats['MVP']:^12} {stats['puntos']:^10}")
print("-------------------------------------------------------")
```
